```python
import csv
import json
import os

import pandas as pd
from pandas.io.json import json_normalize 
```

```python
os.chdir('C:\\Users\\dukho\\projects\\company sentiment\\data')
docs_file = '..\\docs.csv'
docs_headers = ['uuid', 'site','author', 'url', 'ord_in_thread', 'title','highlightText', 'language','text','published', 'crawled', 'highlightTitle']
orgs_file = '..\\orgs.csv'
orgs_headers = ['uuid','organization','sentiment']
```

### Creating csv with headers for 1st file with docs info

```python
with open(docs_file,'w',encoding="utf-8",newline="") as csv_doc:
    writer = csv.DictWriter(csv_doc, fieldnames=docs_headers)
    writer.writeheader()
```

### Creating cwv with headers for 2nd file with organizations info

```python
with open(orgs_file,'w',encoding="utf-8",newline="") as csv_org:
    writer = csv.DictWriter(csv_org, fieldnames=orgs_headers)
    writer.writeheader()
```

### Populating the data frame with document info

```python
%%time

for dirpath, dirnames, files in os.walk('.'):
    for file_name in files:
        with open(os.path.join(dirpath,file_name), encoding="utf-8") as data_file:
            data = json.load(data_file)
            with open(docs_file,'a',encoding="utf-8",newline="") as csv_doc:
                writer = csv.DictWriter(csv_doc, fieldnames=docs_headers)
                writer.writerow({'uuid': data['uuid'],
                         'site': data['thread']['site'],
                         'author': data['author'],
                         'url': data['url'],
                         'ord_in_thread': data['ord_in_thread'],
                         'title': data['title'],
                         'highlightText': data['highlightText'], 
                         'language': data['language'],
                         'text': data['text'],
                         'published': data['published'], 
                         'crawled': data['crawled'], 
                         'highlightTitle': data['highlightTitle']                                      
                        })
        with open (orgs_file,'a',encoding="utf-8",newline="") as csv_org:
            writer_org = csv.DictWriter(csv_org, fieldnames=orgs_headers)
            for org in data['entities']['organizations']:
                writer_org.writerow({'uuid': data['uuid'],
                                     'organization': org['name'],
                                     'sentiment': org['sentiment']
                                    })

```
